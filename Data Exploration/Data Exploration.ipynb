```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('pdb_data_no_dups.csv')
```

```python
data.head()
```

```python
seq_df = pd.read_csv('pdb_data_seq.csv')
```

```python
seq_df = seq_df.drop(['residueCount', 'macromoleculeType'],axis=1)
```

```python
df = pd.merge(data, seq_df, on='structureId', how='inner')
df.head()
```

```python
protein_df = df[df['macromoleculeType']=='Protein']
```

```python
protein_df.head()
```

```python
protein_df.isnull().sum()
```

```python
protein_df.shape
```

```python
protein_df = protein_df.drop(['crystallizationMethod','crystallizationTempK', 'publicationYear', 'resolution'],axis=1)
```

```python
protein_df.isnull().sum()
```

```python
protein_classifications = protein_df[['classification','sequence']]
```

```python
protein_classifications
```

```python
counts =protein_classifications['classification'].value_counts()
```

```python
types = np.asarray(counts[(counts > 1000)].index)
df = protein_classifications[protein_classifications.classification.isin(types)]
```

```python
df.to_csv('protein_df.csv', index=False)
```
